#Multiple linear regression using Sklearn to predict the gpa based on SAT score and a random value of 1 to 3
###The outcome is a feature selection to find which variable is not adding value to the model

##Import libraries and load data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() #set to the default style

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/1.02.+Multiple+linear+regression.csv')
data.head()
```

```python
data.describe()
```

##Creating the miltiple linear regression

###Declaring the dependent and independt variables

```python
x = data[['SAT', 'Rand 1,2,3']]
y = data['GPA']
```

###Regression

```python
reg = LinearRegression()
#we dont need to transform the x value into a matrix as we do in linear regression
reg.fit(x,y)
```

```python
reg.coef_
#the first value is for the SAT, second for Rand 1,2,3
```

```python
reg.intercept_
```

##R-Squared

```python
reg.score(x,y )
```

##Adjusted R-Squared
###To get more accurate measurements

```python
x.shape
```

```python
r2 = reg.score(x,y)
n = x.shape[0] #number of observations
p = x.shape[1] #number of features

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

####As the adjusted R-squared is less than the R-squared, it shows that theres a variable that is not adding value to the model.
####To find out which variable, we use the p value
####p-value > 0.05, we can disregard it  

##Feature selection
####F-regression creates simple linear reg of each feature and the dependant variable.
####For a simple linear regression, the p value of F-Stat is equivilent to the p value of the only independent variable

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
#First array is the F-statistics, the second array is the p values
```

```python
#As we are only interested in the p value, we need to extract it
p_values = f_regression(x,y)[1]
p_values
```

```python
p_values.round(3)
#The first value is SAT, second is Rand 1,2,3
```

####As Rand 1,2,3 had a p-value>0.05 it shows that it is useless and not adding value to the model.
####While the p-value of SAT is 0, it shows that it is highly significant in predicting the SAT score

###the f-regression does NOT take into account the interrelation of the features

##Creating a summary table

```python
reg_summary = pd.DataFrame(data=x.columns.values, columns=['Features'])
reg_summary
```

```python
#Adding 2 columns
reg_summary ['Coefficients'] = reg.coef_
reg_summary ['p-values'] = p_values.round(3)
reg_summary
```
